```python
import os, time
import requests, bs4


DATA_PATH = "data"
BOOK_PATH = os.path.join(DATA_PATH, "books")

SITE_ADDR = "http://lib.pushkinskijdom.ru"
ROOT_SUFFIX = "/Default.aspx?tabid=4863"


def ensure_paths():
    for path_to_ensure in (DATA_PATH, BOOK_PATH):
        if not os.path.exists(path_to_ensure):
            os.mkdir(path_to_ensure)


def url_suffix_to_filename(suffix):
    return "".join(c for c in suffix if c.isalnum()) + ".html"



last_download = time.time()


def download_or_read(url_suffix, subdir, or_read=True, download_period=5.):
    global last_download
    
    file_path = os.path.join(subdir, url_suffix_to_filename(url_suffix))
    if os.path.exists(file_path):
        print(file_path, "exists, skipping")
        if or_read:
            return open(file_path).read()
        else:
            return None
    
    cur_time = time.time()
    delta = cur_time - last_download
    if delta < download_period:
        time.sleep(download_period - delta)
    
    print("to download:", SITE_ADDR + url_suffix)
    resp = requests.get(SITE_ADDR + url_suffix)
    last_download = time.time()
    
    assert resp.status_code == 200, resp.status_code
    print("Writing", file_path)
    open(file_path, "wb").write(resp.content)
    return resp.content


def download_all_pages():
    ensure_paths()
    
    root_page = download_or_read(ROOT_SUFFIX, DATA_PATH)
    root_parsed = bs4.BeautifulSoup(root_page)
    
    pos_book_links = root_parsed.find_all('a', {'class' : 'rtIn'})
    print("Ссылки на странице: (" +  str(len(pos_book_links)) + ")")
    for i in pos_book_links:
        download_or_read(i['href'], BOOK_PATH)
```
